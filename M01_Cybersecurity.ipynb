## Security Management

### Definition of Security Management (Colab AI)

**Security management** is the <font color="red">process</font> of <font color="blue">identifying</font>, <font color="blue">assessing</font>, and <font color="blue">mitigating</font> security <font color="red">risks</font> to an organization's <font color="red">assets</font>. It involves the <font color="blue">implementation</font> of security policies, procedures, and technologies to protect against unauthorized access, use, disclosure, disruption, modification, or destruction of information or property.

Security management typically includes the following steps:


* **Risk assessment**: Identifying and assessing the security risks to an organization's assets.
* **Policy development**: Developing security policies and procedures to mitigate the identified risks.
* **Technology implementation**: Implementing security technologies, such as firewalls, intrusion detection systems, and encryption, to protect against threats.
* **Employee education**: Educating employees about security risks and how to protect against them.
* **Incident response**: Developing and implementing a plan to respond to security incidents.
* **Monitoring and review**: Continuously monitoring the security of an organization's assets and reviewing security policies and procedures to ensure they are effective.

Security management is an ongoing process that requires the cooperation of all employees in an organization. It is essential for protecting an organization's assets and reputation.

### Definition of Safety, Security and Assurance

* What is *safe*?

    * **Confidentiality** measures are designed to prevent sensitive information from unauthorized access attempts.
    * **Integrity** involves maintaining the consistency, accuracy and trustworthiness of data over its entire lifecycle (in use, in transit, at rest).

  * Managers establish security controls.
    * Rules, policies, standards, laws, regulations, ...
* What is *assurance*?
  * Assurance
    * **Auditing** is an examination of the management controls. The evaluation of evidence obtained determines if the information systems are safeguarding assets, maintaining data integrity, and operating effectively to achieve the organization's goals or objectives.
    * **Testing** is a process intended to reveal flaws in the security mechanisms of an information system that protect data and maintain functionality as intended.
    * **Monitoring** involves collecting and analyzing information to detect suspicious behavior or unauthorized system changes, defining which types of behavior should trigger alerts, and taking action on alerts as needed.

* Risk versus Loss versus Cost (tangible and intangible)
  * Risk Avoidance tries to eliminate risk. This strategy is good for risks that could potentially have a significant impact on a business or project.
  * Risk Mitigation may limit the impact of risk so that if a problem occurs it will be easier to fix. It is the most common strategy.
  * Accepting risk occurs when a business or individual acknowledges that the potential loss from a risk is not high enough to warrant spending money to avoid it.
  * The rationale behind risk acceptance is that the costs to mitigate or avoid risks are too great to justify given the small probabilities of a hazard, or the small estimated impact it may have.
  * Self-insurance is a form of risk acceptance.
  * On the other hand, insurance transfers risk to a third party.

## Malware Ontology (Colab AI)

* **Malware**: Any software that is designed to damage or disrupt a computer system.
* **Malware family**: A group of malware that share similar characteristics, such as their functionality, target, or distribution method.
* **Malware variant**: A specific instance of a malware family that may have unique features or characteristics.
* **Malware infection**: The process by which malware is introduced into a computer system.
* **Malware attack**: The use of malware to compromise a computer system or network.



## Malware (Colab AI)

* **Viruses**: Viruses are self-replicating programs that attach themselves to other programs or files. When the infected program or file is executed, the virus code is also executed, allowing the virus to spread to other computers.
* **Worms**: Worms are similar to viruses, but they do not need to attach themselves to other programs or files. Instead, they can spread independently by exploiting vulnerabilities in computer systems.
Trojan horses: Trojan horses are disguised as legitimate programs or files, but they actually contain malicious code that can be executed when the program or file is opened.
* **Ransomware**: Ransomware is a type of malware that encrypts files on a victim's computer and then demands a ransom payment in exchange for the decryption key.
* **Spyware**: Spyware is a type of malware that collects information about a victim's activities on their computer, such as their browsing history, keystrokes, and passwords.
Adware: Adware is a type of malware that displays unwanted advertisements on a victim's computer.
* **Scareware**: Scareware is a type of malware that displays fake security warnings or alerts in order to trick victims into paying for fake security software.
* **Rootkits**: Rootkits are a type of malware that gives attackers privileged access to a victim's computer system.

## Attack Lifecycle


MITRE ATT&CK is organized into a hierarchy of tactics, techniques, and procedures (TTPs).

* Tactics: The high-level goals that an adversary is trying to achieve, such as "Initial Access" or "Execution".
* Techniques: The specific methods that an adversary uses to achieve a tactic, such as "Phishing" or "Exploit Public-Facing Application".
* Procedures: The specific steps that an adversary takes to execute a technique, such as "Send Spearphishing Email" or "Use DLL Side-Loading".

  * 14 tactics
    * Reconnaissance
    * Resource Development
    * Initial Access
    * Execution

    * Privilege Escalation
